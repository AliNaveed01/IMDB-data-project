SINCE THE DATA IS RICH AND WE HAVE A VERY SMALL RATIO OF NULL VALUES IN THE CODE, WE WILL JUST FILL THEM WITH THE MEAN MEDIANS OR EITHER DROP EM


FIRST, WE WILL BE CLEANING THE DATASET WITH 5K ENTRIES

```python
# Read the data from 5k_IMDB_MovieS.xlsx
file = pd.read_excel('5k_IMDB_MovieS.xlsx')
file.drop(columns=['Unnamed: 0'], inplace=True)
file.head()
```

```python
file.isnull().sum(axis=0).sort_values(ascending=False)
```

```python
#fill the null values in the column 'imdbRating' with the mean of the column
file.imdb_score.fillna(file.imdb_score.mean(), inplace=True)
#fill the null values in the column 'gross' with the mean of the column
file['Gross Collection'].fillna(file['Gross Collection'].mean(), inplace=True)
#drop the budget column
#file.drop(columns=['budget'], inplace=True)
#forwards fill the null values in the column 'Year of release'
file['Year of release'].fillna(method='ffill', inplace=True)
file.language = file.language.fillna('English')
file.dropna(subset=['Director'], inplace=True)
#drop the rows where the column 'Director' has null values
file.dropna(subset=['Star'], inplace=True)
#drop the column 'budget'
#file.drop(columns=['budget'], inplace=True)
```

```python
#sort the dataset by the column 'imdbRating'
file.sort_values(by=['imdb_score'], ascending=False, inplace=True)
file.head()
```

```python
file.isnull().sum(axis=0).sort_values(ascending=False)
```

```python
#convert file2 in excel format
file.to_excel('clean_5k.xlsx')
```

NOW WE WILL BE DEALING WITH THE DATASET WITH 1000 VALUES

```python
file2=pd.read_excel('Top_1000_IMDB_Movies.xlsx')
file2.head()
```

```python
#drop the rows where the column 'description' is equal to '*****'
file2.drop(file2[file2.Description == '*****'].index, inplace=True)
#drop the rows where the column 'metascore' is equal to '-----'
file2.drop(file2[file2.Metascore == '-----'].index, inplace=True)   
#drop the rows where the column 'Gross Collection ' is equal to '*****'
#file2.drop(file2[file2['Gross Collection '] == '*****'].index, inplace=True)
file2['Gross collection'].replace('*****', 'NaN', inplace=True)
#drop the column 'Unnamed: 0'
file2.drop(columns=['Unnamed: 0'], inplace=True)
#replace the '---' value in the column 'PG_rating with 'Not Rated'
file2.PG_Rating.replace('---', 'Not Rated', inplace=True)
```

```python
#pd.toexcel('5k_IMDB_MovieS.xlsx', index=False)
#convert file2 into new excel file with the name 'Top_1000_IMDB_Movies_cleaned.xlsx'
file2.to_excel('Top_1000_IMDB_Movies_cleaned.xlsx', index=False)
```
